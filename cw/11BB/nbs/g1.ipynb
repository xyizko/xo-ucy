# Calculating the Gas Price from Explorer 

1. *Transaction Fee* = Gas Used * Gas Price (ETH)



- the above diagram is a screenshot from Seplia Explorer
- 

```python
# Calculating the gas price basd on above diagram 

def calc_gas(gas_price, gas_used):
    gas_price_ineth = gas_price * gas_used 
    return rprint(f"""
gas_price = {gas_price}
gas_used = {gas_used}
-----
TxFee = {gas_price_ineth:.30f}
    """)
    
# Implement function 
calc_gas(0.00000000500000000, 21000 )
```

# SHA 256 Hash Related 
- Getting hashes in various encryption alogirthms

```python
# Function for making hashes

import hashlib

text_convert = b'Upskirt'

def multi_hashes():
    md5_hash = hashlib.md5(text_convert).hexdigest()
    sha256_hash = hashlib.sha256(text_convert).hexdigest()
    sha512_hash = hashlib.sha512(text_convert).hexdigest()
    sha3_256_hash = hashlib.sha3_256(text_convert).hexdigest()
    sha3_512_hash = hashlib.sha3_512(text_convert).hexdigest()

    rprint(f"""

Converting - {text_convert} 
---
md5_hash = {md5_hash} = Characters {len(md5_hash)}
sha256_hash = {sha256_hash} = Characters {len(sha256_hash)}
sha512_hash = {sha512_hash} = Characters {len(sha512_hash)}
sha3_256_hash = {sha3_256_hash} = Characters {len(sha3_256_hash)}
sha3_512_hash = {sha3_512_hash} = Characters {len(sha3_512_hash)}
""")
    

# Execute function 
multi_hashes()
```
